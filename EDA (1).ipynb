```python
q1 =df.describe()["Age"]["25%"]
q1
```

```python
IQR = q3 -q1
```

```python
IQR
```

```python
upper_limit = q3 + 1.5*IQR
upper_limit
```

```python
lower_limit =q1-1.5*IQR
lower_limit
```

```python
# getting outlier
outliers = df[(df["Age"] < lower_limit) | (df["Age"] > upper_limit)]
```

```python
outliers
```

```python
#trimming technique , # first we get a non outlier then that put in original dataframe
```

```python
non_outlier = df[(df["Age"] >= lower_limit) & (df["Age"] <= upper_limit)]
```

```python
non_outlier
```

```python
df = non_outlier #putting non outlier in orignial dataframe from that outlier will get completely remove
```

```python
df
```

```python
#Capping_Technique -- in this we fill the outlier value with Upper limit and lower limit value respectively
df["Age"] = df["Age"].clip( lower_limit , upper_limit)
```

```python
df["Age"] = df["Age"].fillna(age_median_value)
```

Feature is also called as column , variable , dimention

feature scaling is a technique used to standardize or normalize numerical data
so that all the columns have a similar range , this helps machine learning
models learn better and faster

it is a way to make all numbers in your data similar in size so the ml models
words better!

standard deviation - is a number that tells you how spread out the scores are .
it helps you understand how much the individual score differs from the average
score.

step 1 - find the average
step 2 - subtract the average from each person's score
step 3 - square those differences
step 4 - find the average of those squared differences
step 5 - take the square root of that average
